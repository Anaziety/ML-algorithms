```python
#BFS WITH COST 

import networkx as nx
import matplotlib.pyplot as plt
from collections import deque

def bfs_with_cost(graph, start, goal):
    visited = set()
    queue = deque([(start, [start], 0)])  # Each item in the queue is a tuple (node, path, total_cost)

    while queue:
        node, path, total_cost = queue.popleft()

        if node not in visited:
            visited.add(node)

            if node == goal:
                return path

            neighbors = graph.get(node, [])
            for neighbor, actual_cost, heuristic_cost in neighbors:
                if neighbor not in visited:
                    new_path = path + [neighbor]
                    new_cost = total_cost + actual_cost
                    queue.append((neighbor, new_path, new_cost))

    return None  # If no path found

graph = {
    "S": [["A", 5, 0], ["B", 7, 0]],
    "A": [["C", 4, 0], ["G", 10, 0]],
    "B": [["D", 3, 0], ["G", 8, 0]],
    "C": [["G", 6, 0]],
    "D": [["G", 9, 0]],
    "G": []
}

# Run BFS from node "S" to node "G"
source = "S"
goal = "G"
path = bfs_with_cost(graph, source, goal)
if path:
    print("BFS Path from", source, "to", goal, ":", path)
else:
    print("No path found from", source, "to", goal)
```

```python
#DFS WITH COST

def dfs_with_cost(graph, start, goal):
    visited = set()
    stack = [(start, [start], 0)]  # Each item in the stack is a tuple (node, path, total_cost)

    while stack:
        node, path, total_cost = stack.pop()

        if node not in visited:
            visited.add(node)

            if node == goal:
                return path

            neighbors = graph.get(node, [])
            for neighbor, actual_cost, heuristic_cost in reversed(neighbors):  # Reversed to maintain LIFO order
                if neighbor not in visited:
                    new_path = path + [neighbor]
                    new_cost = total_cost + actual_cost
                    stack.append((neighbor, new_path, new_cost))

    return None  

graph = {
    "S": [["A", 5, 0], ["B", 7, 0]],
    "A": [["C", 4, 0], ["G", 10, 0]],
    "B": [["D", 3, 0], ["G", 8, 0]],
    "C": [["G", 6, 0]],
    "D": [["G", 9, 0]],
    "G": []
}

# Run DFS from node "S" to node "G"
path = dfs_with_cost(graph, source, goal)
if path:
    print("DFS Path from", source, "to", goal, ":", path)
else:
    print("No path found from", source, "to", goal)
```

```python
#Anam implemented this A star Algorithm

Anam_Graph = {
    'A': [('B', 6), ('C', 3)],
    'B': [('D', 3), ('E', 2)],
    'C': [('F', 1), ('G', 5)],
    'D': [],
    'E': [],
    'F': [] ,
    'G': [],
}

def get_neighbors(v):
    if v in Anam_Graph:
        return Anam_Graph[v]
    else:
        return None
        
def h(n):
        H_dist = {
            'A': 10,
            'B': 8,
            'C': 5,
            'D': 7,
            'E': 3,
            'F': 6,
            'G': 5,            
        }
        return H_dist[n]

def aStarAlgo(start, stop):
         
        open_set = set(start) 
        closed_set = set()
        g = {} 
        parents = {}
        g[start] = 0
        parents[start] = start
        
        while len(open_set) > 0:
            n = None

            for v in open_set:
                if n == None or g[v] + h(v) < g[n] + h(n):
                    n = v
     
            if n == stop or Anam_Graph[n] == None:
                pass
            else:
                for (m, weight) in get_neighbors(n):
                    if m not in open_set and m not in closed_set:
                        open_set.add(m)
                        parents[m] = n
                        g[m] = g[n] + weight
                         
                    else:
                        if g[m] > g[n] + weight:
                            g[m] = g[n] + weight
                            parents[m] = n
                            if m in closed_set:
                                closed_set.remove(m)
                                open_set.add(m)
 
            if n == None:
                print('Path does not exist!')
                return None
            if n == stop:
                path = []
 
                while parents[n] != n:
                    path.append(n)
                    n = parents[n]
 
                path.append(start)
 
                path.reverse()
 
                print('Path found: {}'.format(path))
                return path
    
            open_set.remove(n)
            closed_set.add(n)
 
        print('Path does not exist!')
        return None

aStarAlgo('A', 'G')
```

```python
#Anam implemented this IDA*

Anam_Graph = {
    'A': [('B', 6), ('C', 3)],
    'B': [('D', 3), ('E', 2)],
    'C': [('F', 1)],
    'D': [],
    'E': [],
    'F': [('G', 5)] ,
    'G': [],
}

def get_neighbors(v):
    if v in Anam_Graph:
        return Anam_Graph[v]
    else:
        return None
        
def h(n):
    H_dist = {
        'A': 10,
        'B': 8,
        'C': 5,
        'D': 7,
        'E': 3,
        'F': 6,
        'G': 5,            
    }
    return H_dist[n]

def ida_star_search(start, stop):
    threshold = h(start)
    path = [start]
    
    while True:
        result, threshold = search(start, stop, 0, threshold, path)
        if result == "FOUND":
            return path
        elif result == float('inf'):
            print("Goal not found.")
            return None

def search(node, goal, g, threshold, path):
    f = g + h(node)
    if f > threshold:
        return f, threshold
    
    if node == goal:
        return "FOUND", threshold
    
    min_threshold = float('inf')
    
    for (neighbor, weight) in get_neighbors(node):
        if neighbor not in path:
            path.append(neighbor)
            result, new_threshold = search(neighbor, goal, g + weight, threshold, path)
            
            if result == "FOUND":
                return "FOUND", threshold
            
            if result < min_threshold:
                min_threshold = result
            
            path.pop()
    
    return min_threshold, min_threshold

start = 'A'
stop = 'G'
print("IDA* Search:")
result = ida_star_search(start, stop)
if result:
    print("Path found:", result)
```
