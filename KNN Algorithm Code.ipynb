```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from collections import Counter


# Load the IRIS dataset
data = pd.read_csv('/content/iris_dataset.csv')
```

```python
data.head()
```

```python
# Separate features and target variable
X = data.drop(columns='target').values
y = data['target'].values
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Function to calculate Euclidean distance
def euclidean_distance(a, b):
    return np.sqrt(np.sum((a - b) ** 2))
```

```python
#k-NN algorithm
def knn(X_train, y_train, X_test_instance, k):
    distances = []
    for i in range(len(X_train)):
        dist = euclidean_distance(X_train[i], X_test_instance)
        distances.append((dist, y_train[i]))
    distances.sort(key=lambda x: x[0])
    knn = [dist[1] for dist in distances[:k]]
    frequency = Counter(knn).most_common(1)[0][0]
    return frequency
```

```python
#predictions and accuracy
def predict(X_train, y_train, X_test, y_test, k):
    correct = 0
    incorrect = 0

    for i in range(len(X_test)):
        prediction = knn(X_train, y_train, X_test[i], k)
        if prediction == y_test[i]:
            correct += 1
        else:
            incorrect += 1

    accuracy = correct / len(y_test)
    print(f"\nAccuracy for k={k}: {accuracy * 100:.2f}%")
    print(f"Correct Predictions: {correct}")
    print(f"Incorrect Predictions: {incorrect}")

# Run the k-NN algorithm for k=7 and k=11
print("Results for k=7:\n")
predict(X_train, y_train, X_test, y_test, 7)

print("\n")

print("Results for k=11:\n")
predict(X_train, y_train, X_test, y_test, 11)
```
